```python
import pandas as pd

data = pd.read_csv('Wholesale_Data.csv')
```

```python
print(data.isnull().sum())
```

```python
print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#  "Fresh"  "Milk" 
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x="Fresh", y="Milk")
plt.show()

# 
```

```python
#  Z-score 
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(data))
outliers = (z_scores > 3).any(axis=1)

# 
print(data[outliers])
```

```python
# 
corr_matrix = data.corr()
print(corr_matrix)

# 
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
#  MinMaxScaler 
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

#  DataFrame
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
```

```python
# 
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=1, max_depth=10)
model.fit(data_scaled,data_scaled)

features = data.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-9:]  # top 10 features

# 
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

# Part II - KMeans Clustering

The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
# 
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# 
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 
scores = []
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_scaled)
    score = silhouette_score(data_scaled, kmeans.labels_)
    scores.append(score)

# 
plt.plot(range(2, 15), scores, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
# k_optimal
k_optimal = scores.index(max(scores)) + 2

# KMeans
kmeans = KMeans(n_clusters=k_optimal)
kmeans.fit(data_scaled)

# 
labels = kmeans.labels_

# 
data['Cluster'] = labels
```

# Part III - Hierarchical Clustering 

Hierarchical clustering is a popular unsupervised machine learning algorithm that is used to identify patterns and group similar data points together in a hierarchy. The algorithm works by iteratively merging or splitting clusters based on a similarity measure until a dendrogram is formed.

To perform hierarchical clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters using techniques such as dendrogram.

```python
# 
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# 
Z = linkage(data, 'ward')

# 
plt.figure(figsize=(12, 8))
dendrogram(Z)
plt.title('Dendrogram')
plt.ylabel('Euclidean distance')
plt.show()

# 
max_d = 50  # 

# fcluster
from scipy.cluster.hierarchy import fcluster
clusters = fcluster(Z, max_d, criterion='distance')

# 
data['cluster'] = clusters

# 
print(data.head())
```

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

```python
# 
from sklearn.decomposition import PCA

# 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# PCA
pca = PCA(n_components=2)

# PCA
data_pca = pca.fit_transform(data_scaled)

# PCADataFrame
data_pca_df = pd.DataFrame(data=data_pca, columns=['PC1', 'PC2'])

# 
print(pca.explained_variance_ratio_)

# 
plt.scatter(data_pca_df['PC1'], data_pca_df['PC2'])
plt.title('PCA on Wholesale Customer Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
